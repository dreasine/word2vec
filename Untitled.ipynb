```python
import jieba
import codecs
from gensim.corpora import WikiCorpus
output = codecs.open('wiki3.txt', 'w',encoding='utf-8')
lists = []
words = []
jieba.set_dictionary('dict.txt.big.txt')
wiki = WikiCorpus('zhwiki-20171103-pages-articles.xml.bz2', lemmatize=False, dictionary={})
print wiki
print wiki.get_texts()
for text in wiki.get_texts():
    #print text
    ci = bytes(' '.join(text).encode('utf-8')).decode('utf-8')
    seg_list = jieba.cut(ci, cut_all=False)
    output.write("".join(text))
    lists.append("".join(seg_list))
    #print (lists)
print ('ok')
```
